```python
import chargement_image as loadData

x_train, y_train = loadData.load_data("./dataset", loadData.CLASSES, image_size=64)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# mlange les donnes charges
shuffler = np.random.permutation(len(y_train))
x_train = x_train[shuffler]
y_train = y_train[shuffler]

plt.figure(figsize=(12, 12))
shuffle_indices = np.random.permutation(9)
for i in range(0, 9):
	plt.subplot(3, 3, i+1)
	image = x_train[shuffle_indices[i]]
	plt.title(loadData.CLASSES[int(y_train[shuffle_indices[i]])])
	plt.imshow(image/255)

plt.tight_layout()
plt.show()
```
